# Departamento de recursos humanos

## Importando librer√≠as

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Cargando Datasets

```python
employee_df = pd.read_csv('data/Human_Resources.csv')
employee_df.head()
```

```python
print(employee_df.info())
```

- 35 caracter√≠sticas en total, cada una con 1470 puntos de datos

```python
employee_df.describe()
```

## Variables categ√≥ricas nominales

```python
employee_df.select_dtypes(include=object).describe()
```

```python
cat_cols = employee_df.select_dtypes(include=object).columns.tolist()
print(cat_cols)
```

```python
cat_df=pd.DataFrame(employee_df[cat_cols].melt(var_name='column', value_name='value')
                    .value_counts()).rename(columns={0: 'count'}).sort_values(by=['column', 'count'])
print(cat_df)
```

Las columnas 'Attrition', 'Over18' y 'OverTime', son preguntas de Yes o No, entonces las dejaremos como variables num√©ricas, donde Yes ser√° 1 y No ser√° 0. El resto de variables ser√°n transformadas a categ√≥ricas

```python
cat_cols = list(set(cat_cols) - set(['Attrition', 'Over18', 'OverTime']))
for cat_col in cat_cols:
    employee_df[cat_col] = employee_df[cat_col].astype("category")
```

```python
employee_df[cat_cols].dtypes
```

## An√°lisis exploratorio de datos

Reemplazamos las columnas 'Attritition', 'Over18' y 'OverTime' por enteros antes de poder llevar a cabo cualquier visualizaci√≥n

```python
employee_df['Attrition'] = employee_df['Attrition'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
employee_df['Over18'] = employee_df['Over18'].apply(lambda x: 1 if x=='Y' else 0)
employee_df['OverTime'] = employee_df['OverTime'].apply(lambda x: 1 if x=='Yes' else 0)
```

Veamos si nos faltan datos, ¬°afortunadamente no es as√≠!

```python
sns.heatmap(employee_df.isnull(), yticklabels=False, cbar=False, cmap='Blues')
plt.show()
```

Algunas caracter√≠sticas como 'MonthlyIncome' y 'TotalWorkingYears' tienen una distribuci√≥n con una cola muy larga (long tail distribution)
Tiene sentido que nos desagamos de 'EmployeeCount' y 'Standardhours' ya que estos campos no cambian de un empleado a otro

```python
employee_df.hist(bins=30, figsize=(20, 20), color='blue')
plt.show()
```

Tiene sentido deshacerse de 'EmployeeCount' , 'Standardhours' y 'Over18' ya que son campos que no cambian de un empleado a otro.
Eliminemos 'EmployeeNumber' tambi√©n.

```python
employee_df.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], axis=1, inplace=True)
```

```python
employee_df
```

Veamos cuantos empleados dejan la empresa!

```python
left = employee_df['Attrition'] == 1
left_df = employee_df[left]
stayed_df = employee_df[~left]
```

Contemos el n√∫mero de empleados que se quedaron y que se fueron.

Parece que estamos ante un conjunto de datos desequilibrado.

```python
print(f'Total = {len(employee_df)}')

print(f'N√∫mero de empleados que dejan la empresa = {len(left_df)}')
print(f'Porcentaje de empleados que dejan la empresa = {round(100.*len(left_df)/len(employee_df), 2)}%')
```

```python
print(f'N√∫mero de empleados que dejan la empresa = {len(stayed_df)}')
print(f'Porcentaje de empleados que dejan la empresa = {round(100.*len(stayed_df)/len(employee_df), 2)}%')
```

```python
left_df.describe()
```

```python
stayed_df.describe()
```

Comparemos la media y el error est√°ndar de los empleados que se quedaron y se fueron
- age': la edad media de los empleados que se quedaron es mayor en comparaci√≥n con los que se fueron
- 'DailyRate': El rate diario de los empleados que se quedaron es mayor
- 'DistanceFromHome': Los empleados que se quedan viven m√°s cerca del trabajo
- 'EnvironmentSatisfaction' & 'JobSatisfaction': Los empleados que se quedan est√°n en general m√°s satisfechos con sus trabajos
- 'StockOptionLevel': Los empleados que se quedan tienen un mayor nivel de stocks options

## Transformaci√≥n de datos

```python
employee_df.head()
```

```python
employee_df.dtypes
```

```python
X_cat_cols = employee_df.select_dtypes(include="category").columns.tolist()
print(X_cat_cols)
```

```python
X_cat = employee_df[X_cat_cols]
```

```python
X_cat.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder(drop='first')
```

```python
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)
X_cat.head()
```

### Columnas num√©ricas

```python
X_num_cols = list(set(employee_df.columns.to_list()) - set(X_cat_cols) - set(['Attrition']))
print(X_num_cols)
print(len(X_num_cols))
```

```python
X_num = employee_df[X_num_cols]
```

```python
X_num.head()
```

### Concatenar DataFrames

```python
X_all = pd.concat([X_cat, X_num], axis=1)
```

```python
X_all.shape
```

## Escalar variables

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(X_all)
```

```python
print(X)
```

```python
print(type(X))
```

```python
y = employee_df['Attrition'].to_numpy()
print(y)
print(type(y))
```

```python
y.shape
```

## Datasets entrenamiento y testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
print(X_train.shape, X_test.shape)
```

## Entrenar y evaluar un classificador utilizando **Regresi√≥n Log√≠stica**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
```

```python
y_pred = model_lr.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(f'Accuracy {round(100*accuracy_score(y_pred, y_test), 2)}')
```

Resultados en el conjunto de testing

```python
cm_lr = confusion_matrix(y_test, y_pred)
sns.heatmap(cm_lr, annot=True)
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```
